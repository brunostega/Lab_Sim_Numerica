# Esercizio 10: Simulated Annealing

Il simulated annealing è un algoritmo di ottimizzazione che sfrutta l'analogia del problema in considerazione con un sistema termodinamico che viene raffreddato lentamente. 
Alla base c'è l'interpretazione della funzione di costo $L(\{x\})$, dove $\{x\}$ è la nostra configurazione, con l'energia di un sistema meccanico statistico con distribuzione di probabilità di Boltzmann
$$
p(\{x\})=\frac{e^{-\beta L(\{x\})}}{Z}
$$

L'idea è quella di eseguire $n$ passi per ogni $beta=\frac{1}{T}$, partendo da temperature alte (bassi valori di $beta$) fino ad arrivare a temperatura nulla, o quasi (alti valori di beta), cambiando poco alla volta la temperatura e facendo un numero di passi sufficiente a far si che il sistema raggiunga l'equilibrio termodinamico, di modo che esso raggiunga la configurazione di minima energia.

Ad ogni passo viene proposta una nuova configurazione $\{x'\}$ che viene accettata con probabilità

$$
p= \begin{cases}  e^{-\beta( L(\{x'\})-L(\{x\})}& \mbox{if }  L(\{x'\})> L(\{x\}) \\ 1& \mbox{altrimenti}
\end{cases}
$$


Per testare l'algoritmo propongo 4 diversi andamenti di $\beta=f(n^{\beta}_{passi})$:
- lineare (spezzata)
- quadratico
- cubico
- esponenziale

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy.optimize as opt
from mpl_toolkits.mplot3d import Axes3D

file=open('output/output.L_circleA.dat')
beta_A, step_A, L_A=np.loadtxt(file, delimiter=",", usecols=[0,1, 2], unpack=True)
file=open('output/output.L_circleB.dat')
beta_B, step_B, L_B=np.loadtxt(file, delimiter=",", usecols=[0,1, 2], unpack=True)
file=open('output/output.L_circleC.dat')
beta_C, step_C, L_C=np.loadtxt(file, delimiter=",", usecols=[0,1, 2], unpack=True)
file=open('output/output.L_circleD.dat')
beta_D, step_D, L_D=np.loadtxt(file, delimiter=",", usecols=[0,1, 2], unpack=True)


steps=[step_A,step_B,step_C,step_D]
betas=[beta_A,beta_B,beta_C,beta_D]
titles=['lineare', 'quadratico', 'cubico', 'esponenziale']

fig, axes=plt.subplots(2, 2, figsize=(15,18))
plt.suptitle(r"Andamenti di $\beta$", fontsize=20, fontweight='bold')
k=0
for i in range(2):
    for j in range(2):
        ax=axes[i][j]
        k
        ax.plot(steps[k],betas[k], '.--')
        ax.set_title(titles[k], fontsize=18, fontweight='bold')
        ax.set_ylabel(r"$\beta$",fontsize=15)
        ax.set_xlabel(r" #step",fontsize=15)
        ax.grid(True)
        k+=1
    


#ax[2].plot(step_B,beta_B, '.--', label="quadratic")
#ax[2].plot(step_C,beta_C, '.--', label="cubic")
#ax[2].plot(step_D,beta_D, '.--', label="exponential")
```

I coefficienti numerici per i vari andamenti sono stati scelti in modo da far si che la temperatura non decrescesse troppo velocemente all'inizio, cioè quell'intervallo dove il sistema è più sensibile a bruschi cambi di temperatura

L'algoritmo viene utilizzato per trovare il percorso più breve per collegare $N_{city}=32$ città (punti), distribuiti casualmente su una circonferenza e all'interno di un quadrato.

In questo caso usiamo $N_{passi}=200$ per ogni valore di $\beta$, con $\beta$ che parte da $\beta_0=0.001$ e arriva a $f(n_{final})$ con $n_{final}=800$.

# Circonferenza

```python
file=open('output/output.circle_pos.dat')
xc,yc=np.loadtxt(file, delimiter=",", usecols=[0,1], unpack=True)

file=open('output/output.order_circleA.dat')
ordercA=np.loadtxt(file, delimiter=",", usecols=[0], unpack=True, dtype="int")
file=open('output/output.order_circleB.dat')
ordercB=np.loadtxt(file, delimiter=",", usecols=[0], unpack=True, dtype="int")
file=open('output/output.order_circleC.dat')
ordercC=np.loadtxt(file, delimiter=",", usecols=[0], unpack=True, dtype="int")
file=open('output/output.order_circleD.dat')
ordercD=np.loadtxt(file, delimiter=",", usecols=[0], unpack=True, dtype="int")


fig, ax=plt.subplots(nrows=4, ncols=3, figsize=(15,18))
ax[0][0].plot(xc,yc, '.')
ax[0][0].set_title("point on a circumference")
ax[0][1].plot(xc,yc, '.-')
ax[0][1].set_title("casual connection")
ax[0][2].plot(xc[ordercA-1],yc[ordercA-1], '.-')
ax[0][2].set_title("connection after alghoritm (LINEAR)")
ax[1][0].plot(xc,yc, '.')
ax[1][0].set_title("point on a circumference")
ax[1][1].plot(xc,yc, '.-')
ax[1][1].set_title("casual connection")
ax[1][2].plot(xc[ordercB-1],yc[ordercB-1], '.-')
ax[1][2].set_title("connection after alghoritm (QUADRATIC)")
ax[2][0].plot(xc,yc, '.')
ax[2][0].set_title("point on a circumference")
ax[2][1].plot(xc,yc, '.-')
ax[2][1].set_title("casual connection")
ax[2][2].plot(xc[ordercC-1],yc[ordercC-1], '.-')
ax[2][2].set_title("connection after alghoritm (CUBIC)")
ax[3][0].plot(xc,yc, '.')
ax[3][0].set_title("point on a circumference")
ax[3][1].plot(xc,yc, '.-')
ax[3][1].set_title("casual connection")
ax[3][2].plot(xc[ordercD-1],yc[ordercD-1], '.-')
ax[3][2].set_title("connection after alghoritm (EXPONENTIAL)")
plt.show()
```

# 10.2: MPI 

Uso la libreria di MPI per parallelizzare il codice di Algoritmo Genetico

Ogni $N_{\text{migration}}$ scambio il miglior individuo in modo casuale tra di diversi core nel seguente modo:

- Genero un numero casuale $A$ tra  $1$ e $N_{\text{core}}-1$, dove $N_{\text{core}}$ è il numero di core utilizzati nella procedura.
- Ogni core $i$ manda il proprio miglior individuo a $PBC(i+A)$ e riceve il miglior individuo da $PBC(i-A)$, dove PBC sono le periodic boundary condition 
$$
PBC(i)=\begin{cases} i, & \mbox{se } 0<i<N_{\text{core}}-1\\ i-A , & \mbox{se } i>N_{\text{core}}-1\\ i+A, & \mbox{se } i<0 \end{cases} 
$$
- Testo il programma parallelizzato nel caso di città distribuite su una circonferenza e su un quadrato utilizzando 4 e 6 core



# $N_{\text{core}}=4$

Uso 6 core e scambio il miglior Individuo ogni $N_{\text{migrazioni}}=30$ generazioni


## Circle

```python
N_ranks=4
ranks=np.arange(1,N_ranks+1,1, dtype=int)

xcs=[]
ycs=[]
orders=[]
for rank in ranks:
    file=open('MPI/Circle/output.circle.rank'+str(rank)+'.pos')
    xc,yc=np.loadtxt(file, delimiter=",", usecols=[0,1], unpack=True)
    xcs.append(xc)
    ycs.append(yc)
    
    file=open('MPI/Circle/output.circle.rank'+str(rank)+'.order.dat')
    order=np.loadtxt(file, delimiter=",", usecols=[0], unpack=True, dtype="int")
    orders.append(order)

fig, ax=plt.subplots(nrows=1, ncols=1, figsize=(8,8))
titles=["core 1","core 2","core 3","core 4"]
for i in range(N_ranks):
    ax.plot(xcs[i][orders[i]-1],ycs[i][orders[i]-1], '.--', label=titles[i])
ax.set_title("Connection of different cores",fontsize=18)
ax.legend()
plt.show()
```

```python
gens=[]
best_Ls=[]
ave_Ls=[]
for rank in ranks:
    file=open('MPI/Circle/output.circle.rank'+str(rank)+'.L.dat')
    gen,best_L, ave_L=np.loadtxt(file, delimiter=",", usecols=[0,1,2], unpack=True)
    gens.append(gen)
    best_Ls.append(best_L)
    ave_Ls.append(ave_L)
    
file=open('../lab9/L_circle.dat')
gen_noMPI,best_L_noMPI, ave_L_noMPI=np.loadtxt(file, delimiter=",", usecols=[0,1,2], unpack=True)


fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(18,10))
for i in range(N_ranks):
    ax[0].plot(gens[i], best_Ls[i], '.--', label=titles[i])
    ax[1].plot(gens[i], ave_Ls[i], '.--', label=titles[i])

ax[0].plot(gen_noMPI, best_L_noMPI,'.-', color='black', label="No MPI")
ax[0].set_title("Best Fitness function of different cores", fontsize=16, fontweight='bold')
ax[0].set_xlabel("# generation", fontsize=15)
ax[0].set_ylabel(r"$L_{best}$", fontsize=15)
ax[0].legend()
ax[0].set_xlim(-10,200)
ax[0].grid(True)

ax[1].plot(gen_noMPI, ave_L_noMPI,'.-', color='black', label="No MPI")
ax[1].set_title("Average Fitness function of different cores", fontsize=16, fontweight='bold')
ax[1].set_xlabel("# generation", fontsize=15)
ax[1].set_ylabel(r"$<L>_{N/2}$", fontsize=15)
ax[1].legend()
ax[1].grid(True)
    
plt.show()

for i in range(N_ranks):
    print("core:", i+1, ", Final fitness linear =", best_Ls[i][len(best_Ls[i])-1])
```

Nella figura a sinistra mostro solo le prime 200 generazioni.

Si osserva come nel caso della circonferenza non vi sia alcun visibile miglioramento con l'utilizzo di 4 core. La funzione di costo decresce in modo simile e anzi, con l'utilizzo dei core sembra addirittura diventare meno efficiente. L'apperente decrescita dell'efficienza, essendo il caso della circonferenza estremamente semplice, è molto probabilmente dovuta ad effetti casuali

## Circle

```python
N_ranks=6
ranks=np.arange(1,N_ranks+1,1, dtype=int)

xcs=[]
ycs=[]
orders=[]
for rank in ranks:
    file=open('MPI6/Circle/output.circle.rank'+str(rank)+'.pos')
    xc,yc=np.loadtxt(file, delimiter=",", usecols=[0,1], unpack=True)
    xcs.append(xc)
    ycs.append(yc)
    
    file=open('MPI6/Circle/output.circle.rank'+str(rank)+'.order.dat')
    order=np.loadtxt(file, delimiter=",", usecols=[0], unpack=True, dtype="int")
    orders.append(order)

fig, ax=plt.subplots(nrows=1, ncols=1, figsize=(10,10))
titles=["core 1","core 2","core 3","core 4","core 5","core 6"]
for i in range(N_ranks):
    ax.plot(xcs[i][orders[i]-1],ycs[i][orders[i]-1], '.--', label=titles[i])
ax.set_title("Connection of different cores",fontsize=18)
ax.legend()
plt.show()
```

```python
gens=[]
best_Ls=[]
ave_Ls=[]
for rank in ranks:
    file=open('MPI6/Circle/output.circle.rank'+str(rank)+'.L.dat')
    gen,best_L, ave_L=np.loadtxt(file, delimiter=",", usecols=[0,1,2], unpack=True)
    gens.append(gen)
    best_Ls.append(best_L)
    ave_Ls.append(ave_L)
    
file=open('../lab9/L_circle.dat')
gen_noMPI,best_L_noMPI, ave_L_noMPI=np.loadtxt(file, delimiter=",", usecols=[0,1,2], unpack=True)


fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(18,10))
for i in range(N_ranks):
    ax[0].plot(gens[i], best_Ls[i], '.--', label=titles[i])
    ax[1].plot(gens[i], ave_Ls[i], '.--', label=titles[i])

ax[0].plot(gen_noMPI, best_L_noMPI,'.-', color='black', label="No MPI")
ax[0].set_title("Best Fitness function of different cores", fontsize=20, fontweight='bold')
ax[0].set_xlabel("# generation", fontsize=15)
ax[0].set_ylabel(r"$L_{best}$", fontsize=15)
ax[0].legend()
ax[0].set_ylim(5,50)
ax[0].set_xlim(-10,200)
ax[0].grid(True)

ax[1].plot(gen_noMPI, ave_L_noMPI,'.-', color='black', label="No MPI")
ax[1].set_title("Average Fitness function of different cores", fontsize=20, fontweight='bold')
ax[1].set_xlabel("# generation", fontsize=15)
ax[1].set_ylabel(r"$<L>_{N/2}$", fontsize=15)
ax[1].legend()
ax[1].set_ylim(5,50)
ax[1].grid(True)
    
    
plt.show()
```

Si osserva come, utilizzando 6 core, sembra esserci un andamento globale della funzione di costo in qualche modo più continuo. Anche in questo caso però non si osserva un rilevante miglioramento con l'utilizzo della parallelizzazione

Aumentando il numero di cores il tempo computazionale diminuisce.

L'utilizzo della parallelizzazione, almeno nel caso del quadrato, sembra quindi migliorare la ricerca del miglior percorso e, all'aumentare del numero di core usati, diminuisce il tempo computazionale. Per migliorare ulteriormente l'algoritmo si potrebbe provare ad aumentare il numero di scambi tra core, sia diminuendo $N_{\text{migrazioni}}$, sia aumentando il numero di miglior individui scambiati ogni $N_{\text{migrazioni}}$.
