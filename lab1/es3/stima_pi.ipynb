# Esercizio 1.3: Stima di pi

Stima di $\pi$ con l'esperimento di Buffon
$$\pi = \frac{2L}{Pd} = \lim_{N_{thr} \to \infty}\frac{2LN_{thr}}{N_{hit}d}$$

utilizzando la statistica a blocchi e usando
- $d=1$
- $L=0.83$
- $N_{thr}=13000$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
file=open('stima_pi.csv')
x,sum_prog,err=np.loadtxt(file, delimiter=",", usecols=[0,1,2], unpack=True)

plt.errorbar(x,sum_prog-np.pi,yerr=err)
plt.xlabel('#throws')
plt.ylabel(r'$<r>-\pi$')
plt.grid(True)
plt.show()
```

Il risultato sembra essere compatibile con il valore esatto di $\pi$
