### Parte 1
Calcolo la media progressiva, e le relative incertezze, in funzione del numero di blocchi di un generatore  di numeri pseudo-casuali uniformemente distribuiti tra [0,1]

Grafico della media progressiva, con relative incertezze del valor medio dei numeri pseudo casuali
calcolato utilizzando la statistica a blocchi

```python
x,sum_prog, err=np.loadtxt('stima_r.csv', delimiter=",", usecols=[0,1,2], unpack=True)

plt.figure(figsize=(14,6))
plt.errorbar(x,sum_prog-0.5,fmt='.--', ecolor='orange',yerr=err)
plt.xlabel('#throws')
plt.ylabel(r'$<r>-1/2$')
plt.title('media progressiva')
plt.grid(True)
plt.show()
```

### Parte 2
Calcolo la varianza progressiva, e le relative incertezze, in funzione del numero di blocchi di un generatore  di numeri pseudo-casuali uniformemente distribuiti tra [0,1]

Grafico della valor medio progressivo della varianza, con relative incertezze del valor medio dei numeri pseudo casuali
calcolato utilizzando la statistica a blocchi

```python
x,sum_prog, err=np.loadtxt('stima_sigma2.csv', delimiter=",", usecols=[0,1,2], unpack=True)

plt.figure(figsize=(14,6))
plt.errorbar(x,sum_prog-1./12,fmt='.--', ecolor='orange',yerr=err)
plt.xlabel('#throws')
plt.ylabel(r'$<(r-0.5)^2>-1/12$')
plt.title(r'varianza progressiva')
#plt.ylim(-0.1,0.1)
plt.grid(True)
plt.show()
```

### Parte 3 
Test del $\chi^2$ per valutare se la distribuzione dei numeri generati pseudo casualmente sia o meno uniforme.
Se $\chi^2 = \sum_{i} \frac{\left( O_i - E_i \right)^2}{E_i} \sim 100$, allora la distribuzione è compatibile con essere uniforme


```python
chi=np.loadtxt('chi2.csv')

print(chi.mean(), "+-", np.sqrt(chi.var()))
```
